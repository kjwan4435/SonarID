### **PreProcessing**

```python
# Upsacaled Zadoff-Chu Sequence Generator

def demodzc(y):
    carrierFreq = 20250.0
    fileFreq = 48000.0 
    x = arange(len(y))
    idown = y * cos(2 * pi * carrierFreq * x / fileFreq)
    qdown = y * (-sin(2 * pi * carrierFreq * x / fileFreq))
    nyq_rate = fileFreq / 2.0   # The Nyquist rate of the signal.
    cutoff_hz = 3500.0
    width = 1000.0
    atten_db = 40.0
    N = round((fileFreq/width) * (atten_db/22))
    taps = signal.firwin(N, cutoff_hz/nyq_rate)

    idown_lp = signal.lfilter(taps, 1.0, idown)
    qdown_lp = signal.lfilter(taps, 1.0, qdown)
    zcseqReceived = idown_lp + 1j*qdown_lp
    
    return zcseqReceived

def zcsequence(u, seq_length):
    zcseq = exp((-1j * pi * u * arange(seq_length) *
                 (arange(seq_length)+1)) / seq_length)
    return zcseq

def normalize(data):
    return (data - np.min(data))/np.ptp(data)  # simple 0-1 normaliztion.

fileFreq = 48000.0
zcseq = zcsequence(63, 127)
zcfft = fft.fft(zcseq)
zcfftPadded = concatenate(
    (pad(array(zcfft[:64]), (0, 897), 'constant'), array(zcfft[64:])), axis=0)
zcseqUpscaled = fft.ifft(zcfftPadded)

def zcseqDivider(n=2):
    result = []
    q = int(len(zcseqUpscaled)/n)
    for i in range(n):
        result.append(np.concatenate((zcseqUpscaled[i*q:], zcseqUpscaled[:i*q])))
    return result

zcseqList = zcseqDivider()

def chunkBySize(lst, n, d):
    for x in range(0, len(lst), n):
        for y in range(d):
            each_chunk = lst[x + int(n/d)*y: n+x +int(n/d)*y]
            if len(each_chunk) < n:
                return
            yield each_chunk
```

```python
def autoStructure():
    retain = 1024
    samplingFrequency, structureData = wav.read(posePATH+poses[0])

    chunkedStructure = list(chunkBySize(structureData, 1024, 2))
    structureSignal = chunkedStructure[100]
    structureDemod = demodzc(structureSignal)
    autoStructure = np.correlate(zcseqList[0], structureDemod, mode="same")
    autoStructure = np.absolute(autoStructure)
    autoStructure = (autoStructure - np.min(autoStructure))/np.ptp(autoStructure)
    allPeaks, props = signal.find_peaks(autoStructure, height=0)    # all the peaks
    NPeaks = 1
    topPeak = allPeaks[np.sort(np.argpartition(
      props['peak_heights'], -NPeaks)[-NPeaks:])][0]   # the top NPeaks peaks
    offset = 512-topPeak
    autoStructure = np.roll(autoStructure, offset)
    result = autoStructure[512-int(round(retain/2)):512+int(round(retain/2))]
    return result
```

```python
def poseACSpect(fnIn):
    samplingFrequency, signalData = wav.read(fnIn)
    zcLen = 1024
    chunkedData = list(chunkBySize(signalData, zcLen, 1))
    signalParts = chunkedData[100:400]
    
    retain = 110  # try 300, which gives a coverage of ~2.1 meters.
    data = np.zeros([len(signalParts), retain])
    data2 = np.zeros([len(signalParts), retain])
    deMod = demodzc(signalParts[25])                         # demodulate

    # cross correlate against original signal
    autoC = np.correlate(zcseqUpscaled, deMod, mode="same")
    autoC = np.absolute(autoC)                              # get abs value
    
    # simple 0-1 normaliztion.
    autoC = (autoC - np.min(autoC))/np.ptp(autoC)

    allPeaks, props = signal.find_peaks(autoC, height=0)    # all the peaks
    NPeaks = 1
    topPeak = allPeaks[np.sort(np.argpartition(
        props['peak_heights'], -NPeaks)[-NPeaks:])][0]   # the top NPeaks peaks
    offset = 512-topPeak

    for i in range(0, len(signalParts)):
        deMod = demodzc(signalParts[i])                         # demodulate
        autoC = np.correlate(zcseqList[0], deMod, mode="same")
        autoC = np.absolute(autoC)                              # get abs value
        # simple 0-1 normaliztion.
        autoC = (autoC - np.min(autoC))/np.ptp(autoC)
        autoC = np.roll(autoC, offset)
        data[i] = autoC[412:522]
        data2[i] = autoC[412:522]
        if (i>0): 
            data2[i] = data2[i] - data2[0]
    data2[0]=0

    data = np.array(data).T
    data = np.atleast_3d(data)
    data2 = np.array(data2).T
    data2 = np.atleast_3d(data2)


    return (data, data2)
```

```python
def plotACSpect(fnIn, i_df, df):
    
    fn = fnIn
    samplingFrequency, signalData = wav.read(fnIn)
    
    zcLen = 1024
    chunkedData = list(chunkBySize(signalData, zcLen, 2))
    
    if (df.correctDown.loc[i_df] == True and df.downTime.loc[i_df] != -1):
        down = int((df.ptsDownT.loc[i_df]- df.startTime.loc[i_df])/(df.endTime.loc[i_df]-df.startTime.loc[i_df])*len(chunkedData))
        up = int((df.ptsUpT.loc[i_df]- df.startTime.loc[i_df])/(df.endTime.loc[i_df]-df.startTime.loc[i_df])*len(chunkedData))
        downIndex = int((down+up)/2)
        signalParts = chunkedData[downIndex-60:downIndex+40]
    else:
        signalParts = chunkedData[0:2]
    
    # how much of the ZC signal do we want to plot? Each step is 0.72 cm.
    retain = 300  # try 300, which gives a coverage of ~2.1 meters.
    data = np.zeros([len(signalParts), retain])
    data2 = np.zeros([int(len(signalParts)/2), retain])
    # get the biggest peak in the first chunk for an offset
    deMod = demodzc(signalParts[0])                         # demodulate

    # cross correlate against original signal
    autoC = np.correlate(zcseqUpscaled, deMod, mode="same")
    autoC = np.absolute(autoC)                              # get abs value
    
    # simple 0-1 normaliztion.
    autoC = (autoC - np.min(autoC))/np.ptp(autoC)

    allPeaks, props = signal.find_peaks(autoC, height=0)    # all the peaks
    NPeaks = 1
    topPeak = allPeaks[np.sort(np.argpartition(
        props['peak_heights'], -NPeaks)[-NPeaks:])][0]   # the top NPeaks peaks
    offset = 512-topPeak
    # end peak
    
    # run analysis of each signalPart, aggregating the resulting arrays in the data2 array.
    for i in range(0, len(signalParts)):

        deMod = demodzc(signalParts[i])                         # demodulate
        # cross correlate against original signal
        autoC = np.correlate(zcseqList[i%2], deMod, mode="same")
        autoC = np.absolute(autoC)                              # get abs value
        # simple 0-1 normaliztion.
        autoC = (autoC)/150000.0
        autoC = np.roll(autoC, offset)
        data[i] = autoC[512-int(round(retain/2)):512+int(round(retain/2))]
        if (i%2==0):
            data2[int(i/2)] = autoC[512-int(round(retain/2)):512+int(round(retain/2))]

    
    data = np.array(data)
    data = np.atleast_3d(data)
    data2 = np.array(data2)
    data2 = np.atleast_3d(data2)

    return(data, data2)

def imag_generator(df,folder):
    result1 = []
    result2 = []
    for i in range(0,len(df)):
        if i<10:
            audio_file = PATH + "sub" +folder + '/trial000' + str(i) + '.wav'
        elif i<100:
            audio_file = PATH + "sub" +folder + '/trial00' + str(i) + '.wav'
        elif i>=100:
            audio_file = PATH + "sub" +folder + '/trial0' + str(i) + '.wav'
        data, data2 = plotACSpect(audio_file, i, df)
        result1.append(data)
        result2.append(data2)
    
    # result1: upscaled autocorrelation image, result2: non-upscaled autocorrelation image
    return result1, result2
```

### **INDIVIDUAL MODEL TEST**

```python
def load_CNN_dataset(df, image):
    label_check = 3
    
    X = []
    y = []
    
    for i in range(0, len(df)):
        label = df.finger.loc[i]

        if (df.correctDown.loc[i]):
            X.append(image[i][:,0:300])
            y.append(label)

    X = np.array(X)
    y = np.array(y)
    y = pd.DataFrame(y)

    n_split = 5
    skf = StratifiedKFold(n_splits=n_split, shuffle=True, random_state=1)

    X_train = []
    X_test = []
    y_train = []
    y_test = []

    for train_index, test_index in skf.split(X, y):
        X_train.append(X[train_index])
        X_test.append(X[test_index])
        y_train.append(y.loc[train_index])
        y_test.append(y.loc[test_index])
      
    for i in range(n_split):
        y_train[i] = tf.keras.utils.to_categorical(y_train[i])
        y_test[i] = tf.keras.utils.to_categorical(y_test[i])

    return X_train, y_train, X_test, y_test
```

```python
def prepare_model():
    mainInput = keras.Input(shape=np.array([100,300,1]))

    x = Conv2D(32,(3,3),(1,1),padding="same",activation="relu")(mainInput)
    x = Conv2D(32,(3,3),(1,1),padding="same",activation="relu")(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    x = Conv2D(64,(3,3),(1,1),padding="same",activation="relu")(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    x = Conv2D(128,(3,3),(1,1),padding="same",activation="relu")(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    x = Flatten()(x)
    x = Dense(128, activation="relu")(x)
    out = Dense(3, activation="softmax")(x)

    model = Model(inputs=[mainInput], outputs=out)
    model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
    return model
```

### **GENERAL MODEL**

```python
def load_CNN_merge_dataset(df, image):
    
    X = []
    y = []

    for j in range(len(df)):
    for i in range(0, len(df[j])):
        label = df[j].finger[df[j].index==i].loc[i]
        if (df[j].correctDown[df[j].index==i].loc[i]):        
            X.append(image[j][i])
            y.append(label)


    X = np.array(X)
    y = np.array(y)
    y = tf.keras.utils.to_categorical(y)
    y = pd.DataFrame(y)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1,  stratify=y)

    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    return X_train, y_train, X_test, y_test
```

```python
def prepare_general_model():
    mainInput = keras.Input(shape=np.array([100,300,1]))

    x = Conv2D(32,(3,3),(1,1),padding="same",activation="relu")(mainInput)
    x = Conv2D(32,(3,3),(1,1),padding="same",activation="relu")(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    x = Conv2D(64,(3,3),(1,1),padding="same",activation="relu")(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    x = Conv2D(128,(3,3),(1,1),padding="same",activation="relu")(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    x = Flatten()(x)
    x = Dense(128, activation="relu")(x)
    out = Dense(3, activation="softmax")(x)

    model = Model(inputs=[mainInput], outputs=out)
    model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
    return model
```

```python
def run_general_model(df_sub, result_sub):
    x_train, y_train, x_test, y_test= load_CNN_merge_dataset(df_sub, result_sub)
    batch_size = 128 
    num_classes = 3 
    epochs = 40 

    model = prepare_general_model()
    model.fit([x_train], y_train, batch_size=batch_size, epochs=epochs, verbose=1, validation_split=0.2)
    score = model.evaluate([x_test], y_test, verbose=0)
    y_pred = model.predict([x_test])
    confMatrix = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))
    print(f'Result:', score)
    print(confMatrix)

    return score, confMatrix
```
